## Targeting Customers

###  There are two types of super bock customers: 
<ul>
<li> Customers who has not bought any heineken product</li>
<li> Customers who buys both brands</li>
</ul>

We deal with the first type first. Our goal is to get them start buying Heineken, thus increase heineken sales

Look at what products heineken customers are buying, and develope a single general profile/taste that represents all heineken customers. Then for all customers who buy super bock exclusively, compute jaccard similarity between the customer and the heineken profile. We only target customers who are similar to the profile.

We tried to implement a user-based knn recommendation system, and look at similarity of each superbock customer and each heineken customer. The computation was too intensive, howeer, therefore we had to make a generalized profile for all Heineken customers instead.

```python
# Source: Saimadhu Polamuri, FIVE MOST POPULAR SIMILARITY MEASURES IMPLEMENTATION IN PYTHON
# url: http://dataaspirant.com/2015/04/11/five-most-popular-similarity-measures-implementation-in-python/
def jaccard_similarity(x,y):
 
    intersection_cardinality = len(set.intersection(*[set(x), set(y)]))
    union_cardinality = len(set.union(*[set(x), set(y)]))
    return intersection_cardinality/float(union_cardinality)
```

```python
heineken_profile = heineken_customer_transactions.prod_id.value_counts().index[1:10000]
```

For super bock customers who has not bought any heineken product, compare how similar they are to heineken buyers. We can target the ones with high similarity to the heineken profile so that they start buying heneken products, thus boosting heineken sales.

```python
similarity_list = []

#find super_bock customer who has not bought heineken
only_bock_customer = bock_customers[~pd.Series(bock_customers).isin(pd.Series(heineken_customers))]

for bock_customer in only_bock_customer:
    #if np.where(only_bock_customer==bock_customer)[0][0] % 100 ==0 :
        #print(np.where(only_bock_customer==bock_customer)[0][0])
    bock_customer_index = transaction_data.index[transaction_data.cust_id == bock_customer].tolist()
    bock_customer_df = transaction_data.iloc[bock_customer_index]
    bock_customer_purchases = bock_customer_df.prod_id
    similarity = jaccard_similarity(bock_customer_purchases, heineken_profile)
    similarity_list = similarity_list + [similarity] 
similarity_list= np.array(similarity_list)
```

See how similar people who buys both brands are to average heineken buyer profile and use it as a reference.

```python
bh_similarity_list = []
for bock_customer in bock_heineken_customers:
    #if np.where(bock_heineken_customers==bock_customer)[0][0] % 100 ==0 :
     #   print(np.where(bock_heineken_customers==bock_customer)[0][0])
    bock_customer_index = transaction_data.index[transaction_data.cust_id == bock_customer].tolist()
    bock_customer_df = transaction_data.iloc[bock_customer_index]
    bock_customer_purchases = bock_customer_df.prod_id
    similarity = jaccard_similarity(bock_customer_purchases, heineken_profile)
    bh_similarity_list = bh_similarity_list + [similarity] 
bh_similarity_list= np.array(bh_similarity_list)
```

## For people who buys product from two brands at the same time:
<ul>
<li> if only buys heineken at full price, ignore</li>
<li> if buys more quantity when heineken product is on discount, send offer and incentive them to buy more, there fore increase heineken sales</li>
</ul>

If a customer has not bought any heineken product with a discount, we ignore them because they either
don't buy beers often or don't care about discount. Of course if we have more time and information we 
 can take a closer look to those profiles, but as of now we want to focus on those who has taken discouts
and study how discounts affect their decisions.

```python

discount_number = bock_heineken_purchase.groupby(['cust_id'])['tran_prod_offer_cts'].sum()
potential_target_customer = discount_number.index[discount_number>0]
no_promotion = discount_number.index[discount_number==0]
```

```python
full_price_quantities = np.array(heineken_transactions.loc[heineken_transactions.index[
    heineken_transactions.cust_id.isin(no_promotion)]].tran_prod_sale_qty)
```

Distribution of quantity purchased at full price(customers who only buys heineken at full price)

```python
print('0 percentile: '+str(np.percentile(full_price_quantities,0)))
print('25th percentile: '+str(np.percentile(full_price_quantities,25)))
print('50th percentile: '+str(np.percentile(full_price_quantities,50)))
print('75th percentile: '+str(np.percentile(full_price_quantities,75)))
print('100th percentile: '+str(np.percentile(full_price_quantities,100)))
```

For all other customers who buy both Heineken and Super bock and has taken discounts, find out who buys more when heineken is on sale and target them because they are more sensitive to price

```python
# For each customer in this group, get their heineken purchase history. If they only
# buy at discount, we target them and send offer to boost sales. If they buy when heineken
# is on sale as well as at full price, we only target them if they buy more quantity at discount
# than at full price in totall.
target_sh_customer = np.array([])
for customer in potential_target_customer :
    heineken_purchase = heineken_transactions.loc[
        heineken_transactions.index[heineken_transactions.cust_id==customer].tolist()]
    heineken_purchase = heineken_purchase.groupby(['tran_prod_offer_cts'])['tran_prod_sale_qty'].sum()
    if sum(np.array(heineken_purchase.index)>0)==len(np.array(heineken_purchase.index)) :
        target_sh_customer = np.append(target_sh_customer,customer)
    else :
        full_price_index = np.argmin(np.array(heineken_purchase.index))
        full_price_quantity = heineken_purchase.iloc[full_price_index]
        if (float(full_price_quantity/sum(heineken_purchase))<0.5) :
            target_sh_customer = np.append(target_sh_customer,customer)
```

After we figure out who we should target, we pull their heineken purchases and send them offers on these products to incentivizing them to buy more

```python
recommend_sh_list = np.array([])
for sh_customer in target_sh_customer :
    target_customer_index = transaction_data.index[transaction_data.cust_id == sh_customer].tolist()
    target_customer_df = transaction_data.iloc[target_customer_index]
    target_customer_purchase  = target_customer_df.loc[target_customer_df.index[
    target_customer_df.prod_id.isin(heineken.prod_id)].tolist()]
    
    purchase_list= target_customer_purchase.groupby(['prod_id'])['tran_prod_sale_qty'].sum().sort_values(ascending = False)
    recommend_id = purchase_list.index[0]
    recommend_sh_list = np.append(recommend_sh_list,recommend_id)
    if len(recommend_list) % 100 == 0 :
        print(len(recommend_list))
```

```python
sh_recommend_df = pd.DataFrame({'cust_id' : target_sh_customer.astype(int),
                                'recommended_heineken_product' : recommend_sh_list.astype(int)})
```
